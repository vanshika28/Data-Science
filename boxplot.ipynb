```python
import pandas as pd
import numpy as np
path="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv"
df=pd.read_csv(path)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
%matplotlib inline
```

```python
sns.boxplot(x="body-style" , y="price" , data=df )
print("We see that the distributions of price between the different body-style categories have a significant overlap, and so body-style would not be a good predictor of price")
```

```python
sns.boxplot(x="engine-location" , y="price", data=df)
print("we see that the distribution of price between these two engine-location categories, front and rear, are distinct enough to take engine-location as a potential good predictor of price.")
```

```python
sns.boxplot(x="drive-wheels", y="price", data=df)
print(" we see that the distribution of price between the different drive-wheels categories differs; as such drive-wheels could potentially be a predictor of price.")
```
